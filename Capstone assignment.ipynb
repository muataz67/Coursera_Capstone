### Question 1

Importing the data into a data frame

```python
import pandas as pd
#import the data from wikipedia
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
# show the dataframe 
df.head(12)
```

Removing cells that with Borough Not assigned 

```python
# drop 'Not assigned' data 
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace = True) 
# show the dataframe 
df.head(12)
```

Combinging Neighbourhoods with same Postal Code

```python
df['Neighbourhood'] = df.groupby(['Postal Code'])['Neighbourhood'].transform(lambda x : ','.join(x)) 
  
# drop duplicate data 
df = df.drop_duplicates()    
  
# show the dataframe 
df.head(12)
```

Assigned the borough name to neighbourhoods with Not assigned using replace

```python
# Replace any Not assigned in the Neighbourhood column with that in the Borough
df_updated = df.replace(df['Neighbourhood'] =='Not assigned', value = df['Borough'] , regex = True) 
```
