### Question 1

Importing the data into a data frame

```python
import pandas as pd
#import the data from wikipedia
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
# show the dataframe 
df.head(12)
```

Removing cells that with Borough Not assigned 

```python
# drop 'Not assigned' data 
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace = True) 
# show the dataframe 
df.head(12)
```

Combinging Neighbourhoods with same Postal Code

```python
df['Neighbourhood'] = df.groupby(['Postal Code'])['Neighbourhood'].transform(lambda x : ','.join(x)) 
  
# drop duplicate data 
df = df.drop_duplicates()    
  
# show the dataframe 
df.head(12)
```

Assigned the borough name to neighbourhoods with Not assigned using replace

```python
# Replace any Not assigned in the Neighbourhood column with that in the Borough
df_updated = df.replace(df['Neighbourhood'] =='Not assigned', value = df['Borough'] , regex = True) 
```

```python
# show the dataframe 
df_updated.head(12)
```

### Question 2

Read in the data from the csv file 

```python
#Read in the file after locally downloading it
location_data = pd.read_csv("Geospatial_Coordinates.csv")  
```

```python
# Look at the data
location_data.head(12)
```

Use Pandas to merge the data for matching Post Codes

```python
#set the Postal code as the index for both dataframes
df_updated = df_updated.set_index('Postal Code')
location_data = location_data.set_index('Postal Code')
```

```python
#perform the merge and assign to a new dataframe
mergedDf = df_updated.merge(location_data, left_index=True, right_index=True)
```

```python
#Look at the merged data
mergedDf.head(12)
```

### Question 3

```python
#Import all the needed libraries
from sklearn.cluster import KMeans # import k-means from clustering stage
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
import folium # map rendering library
import numpy as np # library to handle data in a vectorized manner
print('Libraries imported.')
```

Use geopy library to get the latitude and longitude values of Toronto City.

```python
#Set the address and obrain the coordinates
address = 'Toronto, Ca'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

Show all the neighbourhoods on the map

```python
# create map of Totonto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(mergedDf['Latitude'], mergedDf['Longitude'], mergedDf['Borough'], mergedDf['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

To use kclustering we will group all the neighbourhoods by Boroughs and split them into 5 clusters

```python
#Group all neighbourhoods by Borough
toronto_grouped = mergedDf.groupby('Borough').mean().reset_index()
toronto_grouped
```

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Borough', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# Add clustering labels
toronto_grouped.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
#Remove the cluster coordinates
toronto_grouped.drop(['Latitude','Longitude'], axis = 1, inplace = True) 
toronto_grouped
```

```python
#Make a copy
toronto_merged = mergedDf

#Assign the cluster labels to each Neighbourhood
toronto_merged = toronto_merged.join(toronto_grouped.set_index('Borough'), on='Borough')
```

```python
#Final check
toronto_merged.head(12) 
```

Create a map showing the neighbourhoods clustered in 5 groups

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
